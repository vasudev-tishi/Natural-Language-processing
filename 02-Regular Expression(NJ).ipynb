#to find the pattern

```python
text = "my phone number is 408-555-1234"
```

```python
phone = re.search(r'\d\d\d-' , text)
phone.group()
```

```python
for phone in re.finditer(r'\d\d\d-' , text):
    print(phone.group())
```

```python
phone_pp = re.compile(r'(\d{3})-(\d{3})-(\d{4})')
re.search(phone_pp,text).group()
```

```python
print(re.search(phone_pp,text).group(1))
print(re.search(phone_pp,text).group(2))
print(re.search(phone_pp,text).group(3))
```

#exclusion

to eclude char , we can use the ^ symbol is conjunction with a set of brackets [] . anything inside thr [] , is excluded

```python
pharse =" therr are 3 number 34 inside 5 this sentence "
```

```python
re.findall(r'[^\d]' , pharse)
```

```python
re.findall(r'[^\d]+' , pharse)
```

```python
test_phrse = ' this is a string ! but it has punctuation . how can we remove it?'
```

```python
re.findall(r'[^!.? ]+' , test_phrse)
```

```python
re.findall(r'[^!.?]+' , test_phrse)
```

```python
clean = " ".join(re.findall(r'[^!.? ]+' , test_phrse))
clean
```
